```python
import pandas as pd
import numpy as np
```

```python
ab = pd.read_csv("bikedata.csv")
ab                                                                                                                            cxzzzxc
```

```python
num_col = ['temp','hum','windspeed']
#categorical_col = ['season', 'yr', 'mnth', 'holiday','weekday', 'workingday','weathersit']
ab.boxplot(num_col)
```

```python
for x in ['temp']:
    q75,q25 = np.percentile(ab.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    ab.loc[ab[x] < min,x] = np.nan
    ab.loc[ab[x] > max,x] = np.nan
```

```python
ab.isnull().sum()
```

```python
ab = ab.dropna(axis = 0)
ab.isnull().sum()
```
